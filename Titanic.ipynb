```python
import pandas as pd
import warnings
warnings.filterwarnings('ignore')
```

```python
df=pd.read_csv("train.csv")
df
```

```python
#Parch refers to the number of parents/children aboard the Titanic for each passenger.
#sibsp=sibling and Spouses(refer to husbands or wives traveling with the passenger).
df.info()
```

```python
df.set_index("PassengerId" ,inplace=True)
df
```

```python
df["Sex"].value_counts()
```

```python
dfnull=df.isnull().sum()
dfnull
```

```python
#Check missing values percentage in each column to determine if i should fill it or delete it
age_perc=dfnull/len(df)*100
age_perc
```

```python
df.drop("Cabin",axis=1,inplace=True)
df
```

```python
df["Age"].fillna(df["Age"].median(),inplace=True)
df
```

```python
#.mode() returns a series containing the most frequent values for each column and i selected the first value[0] 
df["Embarked"].fillna(df["Embarked"].mode()[0],inplace=True)
df
```

```python
df.isnull().sum()
```

```python
df["familysize"]=df["SibSp"]+df["Parch"]+1
df
```

```python
df.drop(["SibSp","Parch"],axis=1,inplace=True)
df
```

```python
df.drop(["Name","Ticket"],axis=1,inplace=True)
df
```

```python
'''from sklearn.preprocessing import LabelEncoder
label_encoder=LabelEncoder()
df["sex"]=label_encoder.fit_transform(df["Sex"])
df'''
```

```python
#instead of LabelEncoder
df['Sex'] = df['Sex'].map({'male': 0, 'female': 1})
df
```

```python
from sklearn.preprocessing import OneHotEncoder
df=pd.get_dummies(df,columns=["Embarked"])
df
```

```python
x = df.drop('Survived', axis=1)
y=df["Survived"]
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score,classification_report
log_regg=LogisticRegression(max_iter=500)
log_regg.fit(x_train,y_train)
y_predlog=log_regg.predict(x_test)
print("Logistic Regression Accuracy: ", accuracy_score(y_test, y_predlog))
print(classification_report(y_test, y_predlog))
```

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score,classification_report
knn=KNeighborsClassifier()
knn.fit(x_train,y_train)
y_predknn=knn.predict(x_test)
print("K-Nearest Neighbors Accuracy: ", accuracy_score(y_test, y_predknn))
print(classification_report(y_test, y_predknn))
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score, classification_report
dt = DecisionTreeClassifier(random_state=42)
dt.fit(x_train, y_train)
y_preddt = dt.predict(x_test)
print("Decision Tree Accuracy: ", accuracy_score(y_test, y_preddt))
print(classification_report(y_test, y_preddt))
```

```python
from sklearn.naive_bayes import GaussianNB
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix

nb_model = GaussianNB()
nb_model.fit(x_train, y_train)

y_pred_test = nb_model.predict(x_test)

print(" naive_bayes Accuracy:", accuracy_score(y_test, y_pred_test))
print(classification_report(y_test, y_pred_test))
```
